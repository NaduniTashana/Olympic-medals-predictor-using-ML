```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the dataset
dataset = pd.read_csv(r"C:\Users\Tashana\Downloads\archive (3)\IRIS.csv")
dataset.head()
```

```python
dataset.info()
dataset["species_code"] = dataset["species"].astype("category").cat.codes
```

```python
#convert dataset to a numpy array
dataset.values
```

```python
for i in dataset.select_dtypes(include='number').columns:
    sns.histplot(data=dataset,x=i)
    plt.show()
```

```python
dataset = np.array(dataset, dtype=object)
data= dataset[:,:4].astype(float)
target= dataset[:,-1].astype(int)
```

```python
data
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data,test_data,train_target,test_target= train_test_split(data,target,test_size=0.2)
```

```python
print(train_data.shape)
print(test_data.shape)
print(train_target.shape)
print(test_target.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model= KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(train_data,train_target)
```

```python
#evaluating the model
predicted_target= model.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(test_target, predicted_target)
conf_matrix = confusion_matrix(test_target, predicted_target)
class_report = classification_report(test_target, predicted_target)

print(f'Accuracy: {accuracy}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```
